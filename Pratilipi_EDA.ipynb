# Percentage of Pratilipis Fully Read (100% read_percentage)

```python
# Percentage of pratilipis fully read
fully_read = user_interaction[user_interaction['read_percent'] == 100]
percent_fully_read = (len(fully_read) / len(user_interaction)) * 100

print(f"Percentage of pratilipis fully read: {percent_fully_read:.2f}%")
```

# Categories with the Most Interactions

```python
# Merge user interactions with meta data for category information
merged_data = pd.merge(user_interaction, meta_data, on='pratilipi_id', how='left')

# Count interactions per category
category_interactions = merged_data['category_name'].value_counts()

print("Top Categories by Number of Interactions:")
print(category_interactions)
```

```python
category_interactions
```

```python
# Plotting the top categories
category_interactions.head(10).plot(kind='bar', figsize=(10, 8), color='skyblue', edgecolor='black')
plt.title("Top Categories by Number of Interactions")
plt.xlabel("Category")
plt.ylabel("Number of Interactions")
plt.show()
```

# Do Users Prefer Shorter or Longer Stories (Based on reading time)?

```python
# Convert published_at to datetime
meta_data['published_at'] = pd.to_datetime(meta_data['published_at'])
```

```python
# Merge user interactions with meta data to get reading time
merged_data = pd.merge(user_interaction, meta_data, on='pratilipi_id', how='left')

# Group by reading time and calculate average read percentage
reading_time_pref = merged_data.groupby('reading_time')['read_percent'].mean()
```

```python
# Plot preferences based on reading time
plt.figure(figsize=(10, 5))
plt.scatter(reading_time_pref.index, reading_time_pref.values, alpha=0.7, color='green')
plt.title("Read Percentage vs Reading Time")
plt.xlabel("Reading Time (seconds)")
plt.ylabel("Average Read Percentage")
plt.grid()
plt.show()
```

# How Do Interactions Change Over Time?

```python
# Resample interactions over days
interactions_over_time = user_interaction.set_index('updated_at').resample('D')['user_id'].count()

# Plot interactions over time
plt.figure(figsize=(10, 5))
interactions_over_time.plot(color='purple')
plt.title("User Interactions Over Time (Daily)")
plt.xlabel("Date")
plt.ylabel("Number of Interactions")
plt.grid()
plt.show()
```

```python
# Plot number of pratilipis published over time
meta_data.set_index('published_at').resample('ME')['pratilipi_id'].count().plot(figsize=(10, 5), color='green')
plt.title("Pratilipis Published Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Pratilipis")
plt.show()
```
